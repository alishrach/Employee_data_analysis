```python
# Average tenure by department
avg_service_by_department = emp_data.groupby('department')['tenure'].mean()
print(avg_service_by_department)

plt.figure(figsize=(10,8))
sns.barplot(hue=avg_service_by_department.index, y=avg_service_by_department.values)
plt.xlabel("Department")
plt.title("Tenure by Department")
plt.show()
```

-> The operations department have employees with higher tenure service followed by procurement, sales & marketing and technology while legal department has the lowest employees serving in that department.

```python
# Average Training Score by department
avg_service_by_department = emp_data.groupby('department')['avg_training_score'].mean()
print(avg_service_by_department)

plt.figure(figsize=(10,8))
sns.barplot(hue=avg_service_by_department.index, y=avg_service_by_department.values)
plt.xlabel("Department")
plt.title("Avg_Training_Score by Department")
plt.show()
```

-> Analytics and R&D have the highest average training scores whereas HR and Sales&Marketing have the lowest.

```python
# Average tenure by department
prev_rating_by_department = emp_data.groupby('department')['previous_year_rating'].mean()
print(prev_rating_by_department)

plt.figure(figsize=(10,8))
sns.barplot(hue=prev_rating_by_department.index, y=prev_rating_by_department.values)
plt.xlabel("Department")
plt.title("Previous Year Rating by Department")
plt.show()
```

-> R&D and Operations have the highest ratings in the previous year than compared to Sales & Marketing and Technology.

```python
age_bins = [20, 25, 30, 35, 40, 45, 50, 55, 60]
age_labels = ['20-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55', '56-60']
emp_data['age_group']=pd.cut(emp_data['age'], bins=age_bins,labels=age_labels,right=False)
```

```python
# Calculate average length of service by age group
avg_length_of_service_by_age = emp_data.groupby('age_group')['tenure'].mean()
print(avg_length_of_service_by_age)
```

```python
# Bar plot of average length of service by age group
sns.barplot(x=avg_length_of_service_by_age.index, y=avg_length_of_service_by_age.values)
plt.xlabel('Age Group')
plt.ylabel('Average Length of Service (Years)')
plt.title('Average Length of Service by Age Group')
plt.show()
```

-> Through this, we find out that the age groups from 46-60 tend to stay for longer periods in the company than the ones in the age group of 20-35

```python
plt.figure(figsize=(6,8))
sns.heatmap(emp_data[['age','no_of_trainings','previous_year_rating','tenure','avg_training_score']].corr(),annot=True,cmap='coolwarm')
```

-> If age and length of service increases the average training scores decreases( negatively correlated)

-> Length of service and age are highly correlated with each other. (0.64) As the age increases, the tenure increases as well. Also show in the bar plot above.

-> As the age increases, the number of trainings tend to decrease since they are already experienced. There is a negative correlation of -0.089.

-> Previous year rating and  number of training have a negative correlation of -0.053 showing little to no relationship.

-> Other variables have correlations which is close to zero indicating a weak or no linear relationship between them.

-> The only pair with a strong correlation is age and length of service.
